```python
##### Data analysis on NBA overs and unders from 2020 season



# import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler





```

```python
# Load in data 
# https://www.basketball-reference.com/teams/CHI/2019_games.html
# https://classic.sportsbookreview.com/betting-odds/nba-basketball/totals/?date=20200110
# Goal: predict line_total from features


# Features: 
# ts_pct, efg_pct, fg3a_per_fga_pct, fta_per_fga_pct, orb_pct, drb_pct, 
# trb_pct, ast_pct, stl_pct, blk_pct, tov_pct, usg_pct, off_rtg, def_rtg, teams, 
# score, away_ts_pct, away_efg_pct, away_fg3a_per_fga_pct, away_fta_per_fga_pct,
# away_orb_pct, away_drb_pct, away_trb_pct, away_ast_pct, away_stl_pct, away_blk_pct,
# away_tov_pct, away_usg_pct, away_off_rtg, away_def_rtg, away_teams, away_score, date



df = pd.read_csv('2020_nba_overs_and_advanced_stats.csv')
df.head()






```

```python
# add truth columns


df['over_hit'] = np.where(df['total_game_score'] >= df['over_under'], 1, 0)
df['under_hit'] = np.where(df['total_game_score'] < df['over_under'], 1, 0)

df.head()
```

```python
# Setting up model for predicitng game score

fields_to_drop = ['home_team', 'home_score', 'away_team', 'away_score', 'over_hit', 'under_hit', 'date', 'total_game_score', 'line', 'away_line', 'over_under']


#X = df.drop(fields_to_drop, 1)
X = df[['off_rtg', 'def_rtg', 'away_off_rtg', 'away_def_rtg']]
y = df[['home_score', 'away_score']].values

X_train, X_test, y_train, y_test = train_test_split(X, y)


parameters = {'bootstrap': False,
              'min_samples_leaf': 3,
              'n_estimators': 50,
              'min_samples_split': 10,
              'max_features': 'sqrt',
              'max_depth': 6}
model = RandomForestRegressor(**parameters)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)



print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Classification of overs

fields_to_drop = ['home_team', 'away_team', 'over_hit', 'under_hit', 'total_game_score']

X = df.drop(fields_to_drop, 1)
y = df['over_hit']

X_train, X_test, y_train, y_test = train_test_split(X, y)

# scale the data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


regressor = RandomForestClassifier(n_estimators=20, random_state=0)

regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Classification of unders

fields_to_drop = ['home_team', 'away_team', 'over_hit', 'under_hit', 'total_game_score']

#X = df.drop(fields_to_drop, 1)
X = df[['off_rtg', 'def_rtg', 'away_off_rtg', 'away_def_rtg']]
y = df['under_hit']

X_train, X_test, y_train, y_test = train_test_split(X, y)

# scale the data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


regressor = RandomForestClassifier(n_estimators=20, random_state=0)

regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
